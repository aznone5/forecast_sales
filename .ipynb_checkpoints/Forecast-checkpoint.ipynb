## Convert to datetime and create "Month" Column

```python

train['date'] = pd.to_datetime(train['date'])


train['month'] = train['date'].dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
train['month'] = pd.Categorical(train['month'], categories=month_order, ordered=True)

test['date'] = pd.to_datetime(test['date'])


test['month'] = test['date'].dt.month_name()
test['month'] = pd.Categorical(test['month'], categories=month_order, ordered=True)
```

## Graphs

```python
total = train.groupby('country')['num_sold'].sum().reset_index()
sns.barplot(x='country', y='num_sold', data=total)

plt.title('Total Number of Products Sold per Country')
plt.xlabel('Country')
plt.ylabel('Total Num Sold')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

```python
total = train.groupby('month')['num_sold'].sum().reset_index()
sns.lineplot(x='month', y='num_sold', data=total)

plt.title('Total Number of Products Sold per Country')
plt.xlabel('Country')
plt.ylabel('Total Num Sold')

plt.xticks(fontsize=10, rotation = 45)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

```python
total = train.groupby('store')['num_sold'].sum().reset_index()
sns.barplot(x='store', y='num_sold', data=total)

plt.title('Total Number of Products Sold per Store')
plt.xlabel('Store')
plt.ylabel('Total Num Sold')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

```python
total = train.groupby('product')['num_sold'].sum().reset_index()
plt.figure(figsize=(25, 12))
sns.barplot(x='product', y='num_sold', data=total)



plt.title('Total Number of Products Sold per Store', fontsize=20)
plt.xlabel('Store',fontsize = 20)
plt.ylabel('Total Num Sold', fontsize = 20)

plt.xticks(fontsize=15, rotation = -11)
plt.yticks(fontsize=20)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

```python
from pycaret.regression import *
from sklearn.model_selection import train_test_split
```

## Model

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop(['num_sold'],axis = 1),train['num_sold'],  test_size=0.2, random_state=42)
```

```python
data = X_train.copy()
data['num_sold'] = y_train
```

```python
reg = setup(data=data, target='num_sold' ,categorical_features=['country', 'store', 'product', 'month'])
```

```python
best_model = compare_models()
```

```python
tuned_model = tune_model(best_model)
```

```python
predictions = predict_model(tuned_model, data=test)
```

```python
predictions
```

```python
predictions = predictions.rename(columns={'id': 'Unnamed: 0', 'prediction_label': 'num_sold'})
```

## Submission

```python
submission_df = pd.DataFrame({'id': predictions['Unnamed: 0'], 'num_sold': predictions['num_sold']})
submission_df["num_sold"] = submission_df["num_sold"].round().astype('int')
submission_df.to_csv('submission.csv', index=False)
```
